```python
import pandas as pd
```

```python
data = pd.read_csv("fuzzy_ratio.csv").T.to_dict()
```

```python
cX = sorted(list(data[0].keys())[1:], key = lambda x: float(x[1:]))
```

```python
dat = {}
for row in data.values():
	# print(row)
	# break
	n = row['Query']
	del row["Query"]
	dat[n] = row
```

```python
data = pd.DataFrame(dat).reindex(cX).T
data
```

```python
data.to_csv("fuzzy_ratio.csv")
```
